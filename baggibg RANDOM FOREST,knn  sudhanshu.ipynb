```python
grid_param={
    "n_estimators":[5,10,50,100,120,150],
    'criterion':['gini','entropy'],
    'max_depth':range(10),
    'min_samples_leaf':range(10)
}
```

```python
rf = RandomForestClassifier(n_estimators=5)
```

```python
grid_search_rf=GridSearchCV(param_grid=grid_param,cv=10,n_jobs=6,verbose=1,estimator=rf)   # verbose = log
```

```python
grid_search_rf.best_params_
```
